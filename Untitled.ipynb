```python
analytic = pd.read_csv('results/numeric_quantum_well_eigenvalues_by_well_length.csv', index_col=0)
```

```python
ax = analytic.plot.scatter(x='L',y='e')
ax.set_xlabel('a (\AA)')
ax.set_ylabel('E (eV)')
```

```python
filtered = analytic.copy()
```

```python
def around(line):
    L = line['L']
    e = line['e']
    return filtered[(abs(filtered['L']-L) < 1.0) & (abs(filtered['e']-e) < 0.05)].shape[0]
filtered['n'] = filtered.apply(around, axis=1)
```

```python
filtered[filtered['n'] > 1].plot.scatter(x='L',y='e')
```
