```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from algaas_quantum_well import AlGaAsQuantumWell
device = AlGaAsQuantumWell()
```

```python
device.evolve_imaginary(iterations=1e-4).eigenvalues
```

```python
fix, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(18,6))
ax1.plot(device.z_ang, np.abs(device.eigenstates[0]))
ax2.plot(device.z_ang, np.abs(device.eigenstates[1]))
ax3.plot(device.z_ang, np.abs(device.eigenstates[2]))
plt.show()
```
